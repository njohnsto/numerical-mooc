# Travail 2 - Maxwell

## Introduction

Dans ce travail nous allons étudier l'évolution du champ électromagnétique dans un conducteur de section carré (dans le plan (xy)), contenant une cavité supraconductrice, elle aussi de section carrée.
La section du conducteur fait $1.5cm$ sur $1.5cm$, alors que celle du supraconducteur fait $0.5cm$ sur $0.5cm$. Le conducteur est homogène et infini selon l'axe z.

N.B. Tout au long de ce travail nous avons pris la convention de noter les vecteurs en gras. De plus comme nous allons résoudre le problème section par section, on se retrouve avec un problème à 2 dimensions et, par abus de language, nous allons souvent parler de carré et non pas de parallélépipède. 

**Équations de Maxwell**
$$
\left\{
\begin{array}{l} % attention {l} et non {1}
\nabla \times \mathbf{E}=-\partial_t\mathbf{B} \\
\nabla \times \mathbf{B}=\mu \mathbf{J} \\
\mathbf{J}=\sigma \mathbf{E}
\end{array}
\right.
$$

Dans les équations ci-dessus, on s'est placé dans un régime quasi stationnaire et on a donc négligé le courant de déplacement. De plus on a, par la loi d'Ohm: $$\mathbf{J} = \sigma \mathbf{E}$$

Avec: $\sigma= 5.96 \cdot 10^7 \ S \ m^{-1}$ (pour la partie non supraconductrice) et $\mu_0=1.26\cdot 10{-6} \ N \ A^{-2}$

**Conditions initiales**

En $t=0$, le courant électrique et le champ électromagnétique sont nuls partout dans le domaine. On applique alors brutalement un courant surfacique $J=1 \ A \ m^{-2}$ sur la surface extérieure du conducteur, dans le sens horaire.

## Questions

### Question 1

*Montrons qu'en vertue des équations de Maxwell ci-dessus, le champ magnétique obéit à une équation de diffusion*

\begin{array}{rlcl}
&\nabla \times (\nabla \times \mathbf{B})&=&\nabla \times (\mu \mathbf{J})\\
\Leftrightarrow & \nabla \cdot (\nabla \mathbf{B})-\nabla^2\mathbf{B}&=& \mu \nabla \times \mathbf{J}\\
\Leftrightarrow & -\nabla^2 \mathbf{B}&=&\mu \nabla \times (\sigma \mathbf{E}) \\
\Leftrightarrow & -\nabla^2 \mathbf{B}&=&\mu \sigma (- \partial_t \mathbf{B})
\end{array}

Finalement on retrouve bien une équation de diffusion:

\begin{equation}
\frac{\partial \mathbf{B}}{\partial t} = \frac{1}{\mu \sigma} \nabla^2 \mathbf{B}
\end{equation}


### Question 2

*Conditions aux bords pour le champ magnétique*

Nous avons choisi de placer le centre du repère au bord inférieur du grand carré, l'axe z sortant de l'écran, l'axe y allant vers le haut et l'axe x dirigé vers la droite.

Remarquons tout de suite que par symétrie du problème, le champ magnétique ne dépendra pas de $z$. De plus il sera dirigé selon l'axe $z$, $\mathbf{B}=-B \mathbf{u}_z$, mais il sera rentrant. Pour voir cela il faut regarder la loi de Biot & Savart.

$$d\mathbf{B} \propto Id\mathbf{l} \times (\mathbf{r}-\mathbf{r}')$$

Où $d\mathbf{l}=\text{distance infinitésimale de la boucle de courant}$, $\mathbf{r}-\mathbf{r}'=\text{vecteur reliant un point de la boucle de courant avec un point de l'espace où on souhaite mesurer} \ \mathbf{B}$


Regardons maintenant les conditions aux bords du grand carré, ainsi qu'aux bords du petit carré.

$\underline{\text{Bords du supraconducteur}}$

Commençons par les conditions aux bords du supraconducteur. Dans le supraconducteur, on a que $\sigma=\infty$. 

Or 
$$\mathbf{E}=\frac{\mathbf{J}}{\sigma} \quad \Rightarrow \mathbf{E} \rightarrow 0 \quad \text{pour} \quad \sigma = \infty \quad \Rightarrow \quad \partial_t \mathbf{B} = - \nabla \times \mathbf{E}=0$$

Donc $\mathbf{B}$ est constant au cours du temps à l'intérieur du supraconducteur. Or initialement le courant électrique, ainsi que le champ électromagnétique, sont égaux à zéro partout dans le domaine. On a donc que $\mathbf{B}=0$

On a donc une condition de Dirichlet: $\mathbf{B}=0$ pour $-0.25\leq x\leq 0.25$ et $-0.25 \leq y \leq 0.25$

En général un supraconducteur va influencer la champ magnétique qui l'entoure. Cette influence peut se voir dans la déformation des lignes de champ du champ magnétique: si on a un carré (fini), les lignes de champ juste autour de ce carré vont être déformées et vont "entourer" le carré, comme le ferait de l'eau qui évite une pierre par exemple. 

Or dans notre cas, le supraconducteur, ainsi que le conducteur, sont infinis selon l'axe $z$, axe selon lequel se trouvent aussi les lignes de champs de $\mathbf{B}=-B \mathbf{u}_z$. Ceci implique qu'il n'y aura pas de déformation du champ magnétique et donc que le supraconducteur n'influence pas le champ magnétique du conducteur. 

Mais on sait que les composantes tangentielles du champ électrique aux travers de la surface sont continues. 

Considérons le milieu 1 comme étant le milieu supraconducteur et le milieu 2 comme étant le milieu conducteur. 
Soit $\mathbf{n}_{21}$ le vecteur normal à la surface, séparant les deux milieux.

$$\mathbf{n}_{21} \times (\mathbf{E}_2-\mathbf{E}_1)=0$$

Or à l'intérieur du supraconducteur le champ électrique est nul (voir ci-dessus): 
$\mathbf{n}_{21} \times \mathbf{E}_2=0 \ $ et $\mathbf{E}_2$ ($\textbf{qui se trouve dans le conducteur}$) est donné par la relation :
$\mathbf{E}_2 = \frac{\mathbf{J}}{\sigma}=\frac{\nabla \times \mathbf{B}}{\mu \sigma}$

On aura donc:

$$\mathbf{n}_{21} \times (-\partial_y B \ \mathbf{u}_x + \partial_x B \ \mathbf{u}_y) =0$$

Le vecteur normal va être différent en fonction de la face du carré considérée. Commençons par les surfaces inférieures et supérieures qui auront le même vecteur normal à un signe près: $\mathbf{n}_{21}=(0,\pm1,0)$

$$\mp \partial_y B \ \mathbf{u}_z =0$$

Pour les surfaces latérales, on aura : $\pm \partial_x B \ \mathbf{u}_z =0$.

Finalement, $ \partial_z B \ \mathbf{u}_z =0$, car le champ magnétique ne dépend pas de $z$.

On retrouve donc des conditions de Neumann pour le champ magnétique au bord du supraconducteur.

$\underline{\text{Bords externes}}$

On remarque que la composante tangentielle du champ magnetique est discontinue du fait de la présence du courant de surface au bord externe du carré.

De plus on sait qu'à l'instant initial $t=0$ on applique un courant de surface $J = 1A/m^2$. Considérons un des côtés du carré. Prenons par exemple le côté droit (où $\mathbf{J}$ va vers le bas), et choisissons ensuite un cadre dans le plan (Oxz) qui entoure ce côté, décrit par $d\mathbf{l}$ et de côtés $i \quad i=1,...,4$.

$$\iint \nabla \times \mathbf{B} \cdot d\mathbf{S} = \oint \mathbf{B} \cdot d\mathbf{l} = \iint \mu_0 \mathbf{J} \cdot d\mathbf{S}$$

Résolvons le côté droit et gauche de l'égalité séparément:

$$\iint \mu_0 \mathbf{J} \cdot d\mathbf{S} =\mu_0 \iint J_{xz} \delta(x) dxdz = \mu_0\int 1 \cdot dz = \mu_0 L$$

avec J =1 

$$ \oint \mathbf{B} \cdot d\mathbf{l} = \sum_{i=1}^4 \oint\limits_{\partial S_i} \mathbf{B}_i \cdot d\mathbf{l}$$

On aura donc que seule l'intégrale sur le côté du cadre à l'intérieur de notre carré va compter. L'intégrale du côté à l'extérieur sera nulle car il s'agit d'un conducteur et le champ magnétique y est donc nul. De plus on sait que $d\mathbf{B} \propto \mathbf{I} \cdot d\mathbf{l}$, donc les deux intégrales sur les côtés parallèles à l'axe $x$ seront nulle. En effet le champ magnétique y sera perpendiculaire à $d\mathbf{l}$.

Finalement, on a donc:

$$ \oint \mathbf{B} \cdot d\mathbf{l} = \int B_z dz = B L$$

$\mathbf{B}$ ne dépend pas de $z$ par symétrie du problème.

On peut refaire ce calcul pour tous les côtés et on conclut que: $ B = -\mu_0$ en $-0.75<x<0.75$ et $-0.75<y<0.75$

### Question 3

*Calculons maintenant l'évolution au cours du temps du champ magnétique par la méthode d'Euler (forward) dans le temps, et avec une discrétisation centrée d'ordre 2 pour le laplacien.*

Pour faire cela nous allons nous inspirer de la méthode utilisé dans le numerical mooc pour l'équation de la chaleur, qui est aussi une équation de diffusion.


```python
import numpy
from matplotlib import pyplot
%matplotlib inline
from matplotlib import rcParams, cm
rcParams['font.family'] = 'serif'
rcParams['font.size'] = 16
import matplotlib.animation as animation
from IPython.display import HTML
```

Commençons par implémenter les données initiales du problème, ainsi que les paramètres numériques (pas de temps, grille de discrétisation spatiale) dont nous aurons besoin.

```python
L = 0.015     #m
l = 0.005     #m
sigma = 5.96*10**7     #S/m
mu_0 = 1.26*10**(-6)   #N/A^2
J_i = 1                #A/m^2    
t = 1.25*10**(-2)      #s
nx = 121
ny = 121
dt = 2.5*10**(-7)       #s

x = numpy.linspace(0,L,nx) 
y = numpy.linspace(0,L,ny)  

dx = L/(nx-1)                   
dy = L/(ny-1)                   

nt=int(t/dt)        
B = numpy.zeros((ny, nx))


print(dx)             #pour la justification du choix de dt
```

**Justifions le choix de dt:**

Comme nous allons devoir discrétiser le champ magnétique (entre autres), il faut que les solutions soient stables. Pour cela on a vu qu'il y avait une contrainte de stabilité $\sigma_{CFL}$. Dans le cas d'une diffusion, on avait dans le module 04_03 que $\sigma_{CFL} = \frac{1}{\sigma\mu_0} \frac{\Delta t}{(\Delta x)^2} + \frac{1}{\sigma\mu_0} \frac{\Delta t}{(\Delta y)^2} < \frac{1}{2}$.

Or ici $\Delta x = \Delta y$: $$\frac{1}{\sigma\mu_0} \frac{\Delta t}{(\Delta x)^2}< \frac{1}{4}$$

En remplaçant avec nos valeurs de $dt$ et $dx$, on trouve que $0.21<0.25$

**Indices des positions:**

Définissons au préalable deux fonctions qui calculent les indices $i$ et $j$ correspondants aux positions $x$ et $y$, de sorte à pouvoir modifier les positions $x$ et $y$ dans le suite du code.

```python
def ind_x(dx,x):
    i = int(x/dx)
    return i

def ind_y(dy,y):
    j = int(y/dy)
    return j
```

**Discrétisation de l'équation de diffusion et implémentation des conditions aux bords**

Nous avons appliqué la discrétisation dite "forward in time" et "central in space".

$$\frac{B_{i,j}^{n+1}-B_{i,j}^{n}}{\Delta t}=\frac{1}{\mu_0\sigma}(\frac{B_{i+1,j}^n-2B_{i,j}^n+B_{i-1,j}^n}{\Delta x^2}+\frac{B_{i,j+1}^n-2B_{i,j}^n+B_{i,j-1}^n}{\Delta y^2})$$

$$B_{i,j}^{n+1}=B_{i,j}^n+\frac{1}{\mu_0\sigma}(\frac{B_{i+1,j}^n-2B_{i,j}^n+B_{i-1,j}^n}{\Delta x^2}+\frac{B_{i,j+1}^n-2B_{i,j}^n+B_{i,j-1}^n}{\Delta y^2})\Delta t$$

**Calcul de l'évolution du champ magnétique** 

Tel que mentionné plus haut, nous avons appliqué la même méthode que dans la 3e leçon du module 4 du MOOC pour obtenir le champ magnétique dans le conducteur après $1.25 \times 10^{-2}$ secondes. Ce temps a été choisi puisque c'est le plus grand dont nous aurons besoin dans ce travail. Calculer l'évolution de $B$ pour cette valeur permet de ne faire tourner la fonction *bdiff* (voir ci-après) qu'une seule fois, ce qui réduit le temps d'exécution de l'ensemble de ce notebook.

Nous créeons d'abord une fonction *bcond*, qui calcule d'une part les indices correspondants aux quatres coins du supraconducteur (à l'aide des deux fonctions précédentes), afin de pouvoir d'une autre part implémenter les conditions aux bords déduites à la question précédente pour le supraconducteur. 

```python
def bcond(B,x_supra1,y_supra1,x_supra2,y_supra2):
    i=ind_x(dx,x_supra1)
    j=ind_y(dy,y_supra1)
    a=ind_x(dx,x_supra2)
    b=ind_y(dy,y_supra2)     #les 4 indices permettent de redéfinir un nouveau carré ou un rectangle 
    
    B[j-1,i:a+1]=B[j-2,i:a+1]      #condition aux bords du supraconducteur
    B[b+1,i:a+1]=B[b+2,i:a+1]
    B[j:b+1,i-1]=B[j:b+1,i-2]
    B[j:b+1,a+1]=B[j:b+1,a+2]
 
    B[j:b+1,i:a+1]=0                #B dans le supraconducteur
    
    return B
```

Nous utilisons ensuite *bcond* dans le code suivant. 

Nous créeons la fonction principale bdiff, semblable à celle utilisée dans le MOOC, qui est celle qui implémente les conditions aux bords externes du conducteur et calcule les valeurs de $B$ partout au sein de ce conducteur en utilisant le résultat de *bcond*, et qui nous rend d'une part une liste de tableaux *mmatrix*, et de l'autre une matrice *B* qui contient les valeurs du champ magnétique dans le conducteur au temps final souhaité. Ceci sera utile pour la toute dernière question.

Le tableau "mmatrix" est de dimension (nt/100)$\times$nx$\times$ny. Il est rempli avec des valeurs intermédiaires du champ magnétique toutes les 100 itérations (nous nous assurons de cela en vérifiant que l'indice dans la boucle est bien un multiple de 100, avant de pouvoir l'ajouter au tableau), ce qui nous sera utile pour l'animation graphique. La justification de la dimension est d'ailleurs donnée dans le cadre de la question 3.

```python
x_supra1=0.005   #m
y_supra1=0.005   #m

x_supra2=0.01    #m
y_supra2=0.01    #m

def bdiff(B, nt, nx, ny, dx, dy, dt, sigma, mu_0):
    
    mmatrix=numpy.zeros((int(nt/100),nx,ny))
    B[:,0] = -mu_0          #conditions aux bords externes
    B[0,:] = -mu_0
    B[:,-1] = -mu_0
    B[-1, :] = -mu_0
    for n in range(nt):
        Bn = B.copy()

        B[1:-1, 1:-1] = Bn[1:-1,1:-1] + 1/(mu_0*sigma) *\
            (dt/(dy**2) * (Bn[2:,1:-1] - 2*Bn[1:-1,1:-1] + Bn[:-2,1:-1]) +\
             dt/(dx**2) * (Bn[1:-1,2:] - 2*Bn[1:-1,1:-1] + Bn[1:-1,:-2]))
      
        B=bcond(B,x_supra1,y_supra1,x_supra2,y_supra2)      #conditions aux bords internes
        
        if (n/100).is_integer():
            mmatrix[int(n/100)]=B.copy()
    
    return mmatrix,B
```

En faisant tourner ce code avec les paramètres du problème, cela nous rend un tableau, que nous avons appelé "mat", contenant ces valeurs intermédiaires, et la matrice *Bphi* qui nous servira au dernier exercice. Elle contient les valeurs du champs magnétique dans le conducteur à l'instant $1,25*10^{-2}$s.

```python
B=numpy.zeros((ny,nx))
mat, Bphi = bdiff(B.copy(), nt, nx, ny, dx, dy, dt, sigma, mu_0)
```

### Question 6

*Que vaut, à l'instant $t=1.25\cdot 10^{-3}s$, la valeur de la densité de courant éléctrique ${\bf J}$ en $(x,y,z)=(2.5\cdot 10^{-3}{\rm m},2.5\cdot 10^{-3}{\rm m},0 {\rm m})$*

Afin de trouver la valeur de la densité de courant, il faut trouver une relation avec le champ magnétique, car on connait la valeur de ce dernier partout dans le domaine pour un instant donné.

Par les équations de Maxwell, on a: $\nabla \times \mathbf{B} = \mu_0 \mathbf{J}$


$$\mathbf{J} = \frac{1}{\mu_0} \left ( \frac{\partial B_z}{\partial y}\mathbf{u}_x -\frac{\partial B_z}{\partial x}\mathbf{u}_y \right ) $$ 

On va donc devoir discrétiser le champ magnétique, mais ici il s'agit de dérivées premières.
On va donc simplement pouvoir appliquer la méthode d'Euler dans l'espace. 

$$
\left\{
\begin{array}{l} 
J_x=\frac{1}{\mu_0} \frac{B_{i+1,j}-B_{i,j}}{\Delta y} \\
J_y=-\frac{1}{\mu_0} \frac{B_{i,j+1}-B_{i,j}}{\Delta x} \\
J_z=0
\end{array}
\right.
$$

Pour répondre à cette question, nous implémentons les calculs ci-dessus sous forme d'une fonction Jdiff, qui prend les mêmes arguments que notre fonction *bdiff*, en plus des indices spatiaux demandés, qui ont été calculés à la question précédente.  

```python
def Jdiff(B, i, j, nt, dx, dy, dt, sigma, mu_0):
   
    J_x=1/mu_0*(B[j,i+1] - B[j,i])/dy
    J_y=-1/mu_0*(B[j+1,i] - B[j,i])/dx
    J_z=0
    J=numpy.array([J_x, J_y, J_z])
    
    return J
```

On utilise le même tableau qu'à la question précédente pour avoir le bon temps. Notons que, bien que les indices spatiaux et le temps soient les mêmes qu'à la question précédente, nous nous permettons de les réécrire ou de les recalculer pour éviter les problèmes si l'on modifie l'ordre d'exécution des différentes parties de ce travail.

```python
x_6=0.0025      #m
y_6=0.0025      #m
t3= 0.00125      #s

i=ind_x(dx, x_6)
j=ind_y(dy, y_6)

nt3=int(t3/dt)
B = mat[int(nt3/100)]

J=Jdiff(B.copy(), i,j, nt3,dx, dy, dt, sigma, mu_0)
print('[J_x,J_y, J_z] =',J, 'A/m^2')
```

### Question 7

*Que vaut, à l'instant $t=1.25\cdot 10^{-2}s$, la valeur du flux du champ magnétique à travers la section $z=0$ du conducteur?*

Le flux du champ magnétique à travers une surface est définie par  $ \ \Phi = \iint\limits_{S} \mathbf{B} \cdot d\mathbf{S}$.

On va devoir discretiser une intégrale de surface, pour cela on a décidé de revenir à la définition plus intuitif d'une intégrale, qui est "la somme sous la courbe donné par la fonction". Pour chaque point de notre grille de discretisation qui correspond à une une valeur de $B$, on va considérer une surface infinitésimale $dxdy$.

On aura donc: $ \ \Phi = \sum_{i} \mathbf{B}_i \ dxdy$ 

Nous créeons pour cela une nième fonction $phi$. Au sein de cette fonction, nous initialisons une variable sumphi à 0. A chaque itérations, donc en balayant tous les indices x et y entre 1 et nx (ou ny), on ajoute à cette variable la valeur de B locale multipliée par $dx*dy$ (notre élément d'aire). Elle nous rend finalement la somme de toutes ces contributions dans le conducteur. Nous avons décider de négliger les valeurs aux bords du carré externe, car celles-ci ne contribuent pas beaucoup aux résultats finale et qu'en les considérant il faudrait prendre une surface qui "déborderait" du carré.


```python
def phi(B, dx, dy, nx, ny):
    sumphi=0
    for i in range(1,nx):
        for j in range(1,ny):
            sumphi=sumphi+B[j,i]*dx*dy
    return sumphi
        
```

Nous avons cette fois enfin besoin du second élément retourné par *bdiff*, qui nous le rappelons est la matrice Bphi du champ magnétique à l'instant voulu ($t=1.25\cdot 10^{-2}s$). Nous utilisons cette matrice Bphi dans la fonction *phi*, et nous récupérons finalement la variable phi, qui correspond à la valeur du flux recherchée.

```python
sumphi2=phi(Bphi.copy(), dx, dy, nx, ny)
print ('phi =', sumphi2, 'Tm^2')
```
